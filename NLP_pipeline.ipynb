```python
from transformers import pipeline
```

```python
ner = pipeline(model = 'dslim/bert-base-NER-uncased')
```

```python
ner('founder of microsoft is bill gates')
```

```python
next_word = pipeline(model='gpt2')
```

```python
next_word('I grew up in france, so i could speak ')
```

```python
summary = pipeline("summarization",model = 't5-base')
```

```python
para = '''
Natural language processing (NLP) is a machine learning technology that gives computers the ability to interpret,
manipulate, and comprehend human language.
'''
```

```python
summary(para)
```
