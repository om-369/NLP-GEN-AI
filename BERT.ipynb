```python
import numpy as np
import pandas as pd
import gensim
```

```python
df = pd.read_csv('/content/Tweets_Tokyo_2020.csv')
```

```python
df.head()
```

```python
docs = df['text'].to_list()
docs
```

```python
import re
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords = stopwords.words('english')
```

```python
nltk.download('punkt')
```

```python
nltk.download('wordnet')
```

```python
# word_tokenizer = word_tokenize()
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from string import punctuation

def preprocessing_text(docs):
  cleaned_text = []

  for text in docs:

    text = text.lower()
    text = re.sub(r'https\S','',text)
    text = re.sub(r'@[\w\d_]','',text)
    text = re.sub(r'#[\w\d_]*','',text)

    text = text.translate(str.maketrans('','',punctuation))  # removing punctuations
    tokens = word_tokenize(text) # tokenize
    tokens = [token for token in tokens if token not in stopwords] # removing punctuations
    tokens = [token for token in tokens if token.isalpha()]  # removing emojis and ,digits
    tokens  = [lemmatizer.lemmatize(token) for token in tokens] # lemmatize
    cleaned_text.append(tokens)
  return cleaned_text


cleaned_text = preprocessing_text(docs)
```

```python
cleaned_text[1780]
```

```python
# [[doc1],[doc2]]
```

```python
from gensim import corpora
from gensim.models import LdaModel
```

```python
dictionary = corpora.Dictionary(cleaned_text)
```

```python
corpus = [dictionary.doc2bow(sent) for sent in cleaned_text]

lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10,passes=10)
```

```python
lda_model.print_topics()
```

```python
# # Inference on new document
```

```python
document = ["congratulations mirabai for winning silver medal",
            'usa and china winning most medals']
new_cleaned_text = preprocessing_text(document)

new_corpus = [dictionary.doc2bow(sent) for sent in new_cleaned_text]

document[0]
```

```python
lda_model.get_document_topics(new_corpus[0])
```

```python
lda_model.get_document_topics(new_corpus[1])
```

```python
#  hyperparameters ---> lda_model = LdaModel(corpus=corpus,id2word = dictionary,num_topics=10,passes=10)

>>> coprus ---> data
>>> id2word ---> ids of words
>>> num_topics ---> no of topics
>>> passes ---> no of passes through corpus during training
>>> alpha ---> document topics distributions
>>> eta ---> word topics distribution
```
