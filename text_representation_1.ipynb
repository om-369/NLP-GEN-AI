```python
#  TF -IDF  ----> term frequency - inverse document frquency
```

```python
TF = no.of times term t occuring in document/ total no. of terms in documents
IDF = log(no.of documents in corpus / no. of documents in corpuse containing term t)
```

```python
df
```

```python
TF-IDF = TF * IDF
```

```python
IDF = log(3/3)= log(1)=0
```

```python
IDF = log(3/1)= log(3)
```
