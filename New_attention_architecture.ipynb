```python
# kv cache:
it reduces redundent calculations of key and value vector for each word

it stores previously calculated Kv vectors

it requires more memory to store kv vectors---- drawbacks
```

```python
## Multi query attention :
for every head in same key and value are used.

#less memory is required compare to multi-head attention
# accuracy will be less compared to multi-head attention
```

```python
# Group - Query attention

for group of heads same key and value are used.

#middle way between  MHA and MQA (provide balance between MHA and MQA)
# use less memory than MHA and gives better accuracy than MQA
```
