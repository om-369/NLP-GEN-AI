### Summarization

```python
from langchain.document_loaders import PyPDFLoader
from langchain.chains.summarize import load_summarize_chain
```

```python
from langchain import OpenAI
```

```python
llm = OpenAI()
```

```python
loader  = PyPDFLoader("/content/dir/attention-is-all-you-need-Paper.pdf")
docs = loader.load()
```

```python
chain = load_summarize_chain(llm = llm)
```

```python
chain.run(docs)
```

### chain type

```python
# stuff
#  default
```

```python
chain = load_summarize_chain(llm,chain_type = "stuff")
chain.run(docs)
```

```python
### token limit error
```

```python
!pip install tiktoken
```

```python
# mapreduce
```

```python
chain = load_summarize_chain(llm=llm, chain_type="map_reduce")
chain.run(docs)
```

#### "stuff "
- -- pro : all data once, no information loss
          - single llm calling
- -- cons: token limit

#### "map_reduce"
- pro : can handle large documents
      - parallel preprocessing
- cons : llm called mutiple times
      - some information loss
      

#### "refine"
- pro: can handle larger documents
      - less information loss >> good quality of output
cons : llm called multiple times
      - sequential process
